Ex2:  Average of Sequence

Prompt for a sequence of numbers, the last number being zero

Print the average of the numbers

If no data is entered (only a zero was entered), you will generate an error (division by zero). If no data is entered, print "Sequence Length is 0".

Enter a Number (zero to quit): 1

  Enter a Number (zero to quit): 7

  Enter a Number (zero to quit): 9

  Enter a Number (zero to quit): 0

  Average: 5.666666666666667

  Enter a Number (zero to quit): 0
  Sequence Length is 0


```python
# user to enter  sequence of numbers, program will stop asking if user enters 0
arrayOfNumbers = []
while True:
  numberEntered= input("Enter a Number (zero to quit): ")
  numberEntered = int(numberEntered)
  if numberEntered == 0:
    print("User entered 0 to quit")
    break
  arrayOfNumbers.append(numberEntered)
print("Array of numbers entered by user: ", end="")
print(arrayOfNumbers)
if len(arrayOfNumbers) == 0:
  print("Sequence Length is 0")
else:
  average = sum(arrayOfNumbers)/len(arrayOfNumbers)
  print("Averageof all numbers added: {}".format(average))


```

Ex5:  Number of Even Values

Prompt for a sequence of numbers, the last number being zero.

Print the number of even values.

 Enter a Number (zero to quit): 2
  
  Enter a Number (zero to quit): 1
  
  Enter a Number (zero to quit): 4
  
  Enter a Number (zero to quit): 0
  
  Number of Even Values:


```python
# user to enter  sequence of numbers, program will stop asking if user enters 0
evenNumbercounter = 0
while True:
  numberEntered= input("Enter a Number (zero to quit): ")
  numberEntered = int(numberEntered)
  if numberEntered == 0:
    print("User entered 0")
    break
  if numberEntered % 2 == 0:
    evenNumbercounter += 1

print("Number of even values: {}".format(evenNumbercounter))

```

Ex3: Temperatures

You have a file that has temperatures in Fahrenheit, one per line.

Create a program that performs the following:

Defines a function called FahrToCel

Accepts a Fahrenheit Temperature (floating point) as a parameter

Calculates the temperature in Celsius - C = (F - 32) * 5 / 9

Opens the file FTemps.txt file for reading

Opens the file CTemps.txt file for writing

Reads a line from the FTemps.txt file

Calculates the Celsius temperature.

Writes the Celsius temperature to the CTemps.txt file. The Celsius values should be 5 characters wide with 1 decimal digit.

Prints the number of records processed

Copy, paste and save the file below as FTemps.txt



```python
#method to convert Fahrenheit to Celsius
def FahrToCel(fahrenheit):
  print("Fahrenheit value is: {}".format(fahrenheit))
  celsius = (fahrenheit - 32) * 5 / 9
  print("Celsius value is: {}".format(celsius))
  return celsius

numberOfrecordsConverted = 0
#with closes the file at the end after the reading/writing is done
with open('/content/fTemps.txt', 'r') as fTemps, \
     open('/content/cTemps.txt', 'w') as cTemps:

#read line from read Only TextFile ftemps
  for record in fTemps:

    fahrValue= float(record.strip())

# call function to convert
    celciusValue = FahrToCel(fahrValue)
#write converted line in reqd
    cTemps.write(f'{celciusValue:5.1f}\n')

    numberOfrecordsConverted += 1
print("Number of records converted: {}".format(numberOfrecordsConverted))
```
