Ex4: Swap Two Words

Prompt for a string consisting of exactly two words separated by a space.

Print a new string with the first and second word positions swapped (the second word is printed first).

This task should not use loops or an if statement.

Enter a string: Hello World

World Hello

```python
#user to enter two words
user_Entered = input("Enter two words: ")
print("User entered {}".format(user_Entered))

#split two words in two separate variables
words = user_Entered.split()
print("First word is {} and second word is {}".format(words[0], words[1]))

#swap words
words[0], words[1] = words[1], words[0]
print("After swapping first word is {} and second word is {}".format(words[0], words[1]))

#combine two words
new_String = " ".join(words)
print("Swapped string is {}".format(new_String))


```

First and Last Occurrence

Prompt for a string. Determine if the string contains the letter "f"

If the string contains one letter "f", print the index

If the string contains more than one letter "f", print the index of the first and last occurance

If the string does not contain any letter "f", then do not print zero (0)

Example

Enter a string: comfort

3

Enter a string: office

1 2

Enter a string: hello

0

```python
#user to enter a string
user_Entered = input("Enter a string: ")
print("User entered {}".format(user_Entered))

#check accurences
first_Occurence = user_Entered.find("f")
last_Occurence = user_Entered.rfind("f")

if first_Occurence == -1:
  print("0")
elif first_Occurence == last_Occurence:
  print(first_Occurence)
else:
  print("{} {}".format(first_Occurence, last_Occurence))

```

Ex7:

Number of Distinct Elements

Prompt for a string containing integers in ascending order separated by spaces.

Load the values into a list.

Print the number of distinct elements in the list.

Do NOT use the list or string functions or methods for this assignment (except the .split() method).

Do NOT use the for x in y iterator; use for x in range(n)

  Enter Values Separated by Spaces: 1 2 2 3 3 3

  Number of Distinct Elements: 3

```python
#user to enter string containing integers in ascending order separated by spaces

user_Entered = input("Enter Values Separated by Spaces: ")
print("User entered {}".format(user_Entered))

#load values in a list
list = user_Entered.split()
print("List is {}".format(list))

distinct_Elements = []

#check if the element is already in distinct elements,
#if not add element in the list
for element in range(len(list)):
  if list[element] not in distinct_Elements:
    distinct_Elements.append(list[element])

print("Number of Distinct Elements: {}".format(len(distinct_Elements)))

```
