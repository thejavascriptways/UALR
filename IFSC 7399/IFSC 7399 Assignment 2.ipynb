Ex 4 Digits in Acending Order

Given a three-digit integer, print "YES" if its digits are in ascending order left to right, print "NO" otherwise.

Do not try to sort the number. Determine each digit and use "if" statements.

Enter a Number: 123

  YES

  Enter a Number: 132

  NO


```python
#user to enter three digit number
numberEntered = input("Enter a three digit number: ")

#iprint user entered three digit number
print("Three digit number entered : " + str(numberEntered))

#convert string input to int value
numberEntered = int(numberEntered)

#find individual numbers
numberatOnesPlace = numberEntered % 10
numberatTensPlace = (numberEntered // 10) % 10
numberatHundredsPlace = numberEntered // 100

#print all three numbers separately
print("Number at ones place: {}".format(str(numberatOnesPlace)))
print("Number at tens place: {}".format(str(numberatTensPlace)))
print("Number at hundreds place: {}".format(str(numberatHundredsPlace)))


if numberatHundredsPlace < numberatTensPlace and numberatTensPlace < numberatOnesPlace:
  print("YES")
else:
  print("NO")

```

Ex 10 Sort Three Numbers

Prompt for three integers and print them in ascending order.

Enter First Number: 5

Enter Second Number: 3

Enter Third Number: 7

3 5 7

```python
# user to enter  three integers
firstNumber = int(input("Enter the first integer: "))

secondNumber = int(input("Enter the second integer: "))

thirdNumber = int(input("Enter the third integer: "))

print("The numbers entered by user: {} {} {}".format(firstNumber, secondNumber, thirdNumber))

# Logic to find the smallest number first
if firstNumber <= secondNumber and firstNumber <= thirdNumber:
    if secondNumber <= thirdNumber:
        print(firstNumber, secondNumber, thirdNumber)
    else:
        print(firstNumber, thirdNumber, secondNumber)
elif secondNumber <= firstNumber and secondNumber <= thirdNumber:
    if firstNumber <= thirdNumber:
        print(secondNumber, firstNumber, thirdNumber)
    else:
        print(secondNumber, thirdNumber, firstNumber)
else:
    if firstNumber <= secondNumber:
        print(thirdNumber, firstNumber, secondNumber)
    else:
        print(thirdNumber, secondNumber, firstNumber)
```

Ex 3 Factorial

Given an integer N, print the sum of: 1! + 2! + 3! + ... + N!

This problem has a solution with only one loop, so try to discover it.

Do NOT use the math library.

Enter N: 5

Sum of Factorials: 153


```python
#user to enter a number
numberEntered = input("Enter a number: ")

#user entered three digit number
print("Three digit number entered : " + str(numberEntered))

#convert string input to int
numberEntered = int(numberEntered)

#staring number in factorial series
startingNumber = 1
#initial sum
sumOfFactorials = 0

for fact in range(1, numberEntered + 1):
  startingNumber = startingNumber * fact
  sumOfFactorials = sumOfFactorials + startingNumber

#if number entered is 5
# Fact   Starting number
# 1         1.            1
# 2         1*2           2
# 3         1*2*3         6
# 4         1*2*3*4       24
# 5         1*2*3*4*5     120
#Total Sum of Factorials =153

print("Sum of Factorials: {}".format(sumOfFactorials))

```

Ex Ladder

For given integer N â‰¤ 9, print a ladder of N steps.

The Kth step consists of the integers from 1 to K without spaces between them.

You can use the sep= and end= arguments of the print() function.

Hint: You can use a For loop within a For loop.

Enter N: 9
  1
  12
  123
  1234
  12345
  123456
  1234567
  12345678
  123456789

```python
#user to enter levels the ladder
numberEntered = input("Enter the number of levels: ")

#user entered number
print("Number of levels entered : " + str(numberEntered))

#convert string input to int
numberEntered = int(numberEntered)

for level in range(1, numberEntered + 1):
  for subLevel in range(1, level + 1):
    print(subLevel, end="")
  print()

```
